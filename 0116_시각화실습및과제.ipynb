```python
# 여기에 공통과제 2번 코드를 작성해주세요. 주석을 꼼꼼하게 작성하여 왜 이렇게 시각화를 했는지 설명해주시면 더 좋습니다!


# 데이터 로드
colors = df['color']
prices = df['price']
carats = df['carat']

# 각 색상별로 가격의 평균 계산
mean_prices = df.groupby('color')['price'].mean()

# 시각화 개선: 막대 그래프 활용, 적절한 제목 사용
fig, ax = plt.subplots()
fig.suptitle("Average Price by Colors", size=14)

# seaborn의 color_palette 함수를 사용하여 색상 부여
colors = sns.color_palette('Set1', len(mean_prices))

# 막대 그래프로 변경
ax.bar(mean_prices.index, mean_prices, color=colors, alpha=0.7)

ax.set_xlabel("Color")
ax.set_ylabel("Average Price")
ax.set_ylim(0, 6000)

plt.show()
```

향후 계획 및 실행 방안: 특정 제품의 영업이익이 높다면 해당 제품의 마케팅 전략을 강화하거나 생산량을 늘릴 수 있는 방안을 고려할 수 있습니다. 또한, 다양한 변수를 함께 고려하여 영업이익을 최적화하는 전략을 수립할 필요가 있습니다.

```python
# 여기에 시각화 코드 혹은 태블로 public 링크를 첨부해주세요.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df2 = sns.load_dataset("Coffee_Chain_Sales")
except Exception as e:
        csv_path = 'Coffee_Chain_Sales.csv'
        df2 = pd.read_csv(csv_path)
        
# 제품 종류별 영업이익 평균 계산
profit_by_product = df2.groupby('Product')['Profit'].mean().sort_values()

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=profit_by_product.index, y=profit_by_product.values, palette='viridis')
plt.title('Average Operating Profit by Product Type') #제품 종류별 평균 영업이익
plt.xlabel('Product Type') #제품 종류
plt.ylabel('Average Operating Profit') #평균 영업이익
plt.xticks(rotation=45)
plt.show()
```
